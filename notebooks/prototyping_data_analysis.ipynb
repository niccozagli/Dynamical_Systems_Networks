```python
import h5py
import pandas as pd

with h5py.File("stats.h5", "r", swmr=True) as fh:
    dset = fh["stats"]
    dset.refresh()
    fieldnames = [s.decode("utf-8") for s in dset.attrs["fieldnames"]]
    df = pd.DataFrame(dset[...], columns=fieldnames)

with h5py.File("state.h5", "r", swmr=True) as fh:
    dset_state = fh["state"]
    dset_time = fh["time"]
    dset_step = fh["step"]
    dset_state.refresh()
    dset_time.refresh()
    dset_step.refresh()
    state = dset_state[...]
    time = dset_time[...]
    step = dset_step[...]
```

```python
df.plot(x="t",y="x1")
```
